```python
import h2o
h2o.init()
from h2o.automl import H2OAutoML
```

```python
dataset=h2o.import_file(path="../datasets/games.csv", header=1, sep=",")
```

```python
x = ['firstBlood', 'firstTower', 'firstInhibitor', 'firstBaron', 'firstDragon', 'firstRiftHerald', 
    't1_champ1id', 't1_champ2id', 't1_champ3id', 't1_champ4id', 't1_champ5id',
    't1_towerKills', 't1_inhibitorKills', 't1_baronKills', 't1_dragonKills', 't1_riftHeraldKills',
    't2_champ1id', 't2_champ2id', 't2_champ3id', 't2_champ4id', 't2_champ5id',
    't2_towerKills', 't2_inhibitorKills', 't2_baronKills', 't2_dragonKills', 't2_riftHeraldKills']
y="winner"
```

```python
train, test = dataset.split_frame([0.7])
```

```python
train[y]=train[y].asfactor()
test[y]=test[y].asfactor()
```

```python
aml = H2OAutoML(max_models=50, sort_metric="mean_per_class_error", max_runtime_secs=600)
aml.train(x=x, y=y, training_frame=train)
```
